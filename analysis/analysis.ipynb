```python
import csv
with open('rocket-launch.csv', 'r') as file:
    data = csv.reader(file)
    header = next(data)
    values = []
    for row in data:
        values.append(row)

print(header)
```

```python
# rocket was largely vertical during launch, so we can use z-acceleration as a proxy for overall acceleration
# plot z-acceleration over time (values element 0 is time, element 3 is z-acceleration)

import matplotlib.pyplot as plt

time = [(float(row[0]) / 1000) - 109.8 for row in values]
net_z_acceleration = [(float(row[3]) - float(row[6])) / 100 for row in values]

plt.figure(figsize=(30, 8), dpi=200)

plt.plot(time, net_z_acceleration)
plt.xlim(-10, 65)
plt.xticks(range(-10, 65, 5))

plt.xlabel('Time (s)')
plt.ylabel('Net Z-Accel (m/s^2)')
# plt.title('Slam Blunck Net Z-Accel Over Time')

plt.grid()
```
